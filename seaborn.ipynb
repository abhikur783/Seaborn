```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Here are five commonly used plots that can be created using the Seaborn library:

Scatter plot: Scatter plot shows the relationship between two continuous variables. It is useful for visualizing patterns and 
    identifying any outliers.

Line plot: Line plot shows the relationship between two continuous variables as a continuous line. It is useful for showing 
    trends over time.

Bar plot: Bar plot shows the distribution of categorical data. It is useful for comparing the frequency of different categories.
    

Histogram: Histogram shows the distribution of a continuous variable. It is useful for understanding the shape of the data and
    identifying any outliers.

Heatmap: Heatmap is a graphical representation of data where the individual values contained in a matrix are represented as 
 colors. It is useful for visualizing the relationship between two variables in a dataset.

The uses of these plots can vary depending on the specific context and research question. However, in general, scatter plots 
and line plots are useful for visualizing trends and identifying any outliers, bar plots are useful for comparing frequencies 
across categories, histograms are useful for understanding the distribution of a variable, and heatmaps are useful for
visualizing relationships between variables.
```

```python
import seaborn as sns

flights = sns.load_dataset("flights")
```

```python
sns.heatmap(flights.pivot("month", "year", "passengers"), cmap="YlGnBu")
```

```python
This will create a heatmap showing the number of passengers traveling each month in each year:
    Here, we used the pivot() function to reshape the "flights" dataset so that the rows correspond to the months, the columns 
    correspond to the years, and the values correspond to the number of passengers. We also used the cmap parameter to set the 
    color map for the heatmap. In this case, we used the "YlGnBu" color map, which goes from yellow (low values) to green 
    (medium values) to blue (high values).
```
